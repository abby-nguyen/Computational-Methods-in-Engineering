### Unforced Damped Systems Analysis:
With a defined initial position and initial velocity, an unforced spring-mass system will oscillate indefinitely. However, applying a damping force that opposes the motion of the mass will cause the oscillation to dissipate. An overdamped system will slowly bring the harmonic motion to an equilibrium. An underdamped system will also slowly bring the harmonic motion to equilibrium, but it will continue to slightly oscillate. A critically damped system will bring the harmonic motion to equilibrium the quickest. The second-order differential equation was solved using Runge-Kutta methods, and the position and velocity of the unforced damped systems are shown above. The graphical calculations align with the theoretical predictions as each system reached equilibrium in accordance with the predetermined damping coefficient.

### Forced Damped Systems Analysis:
The spring-mass-damper system was studied with four different inputs/force functions were implemented. This force is applied to the mass, and the overdamped, critically damped, and underdamped cases were analyzed.

First, a constant applied force was studied. As expected, the dampening force was still able to bring the system to equilibrium

Second, a consistently changing applied force was studied: one linear and one quadratic. These systems did not reach an equilibrium as dampening a system with an increasing force input is difficult.

Lastly, an oscillating applied force was studied. The models were expected to dissipate, however, the calculated graphs showed that the systems did not dissipate. This could have been a result of slight and accumulative errors in the Runge Kutta method, or a human processing error.
